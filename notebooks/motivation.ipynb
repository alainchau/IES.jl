# Simplex rectangle example

This notebook illustrates a deficiency of Diffusion Maps.

```julia
W = 6
H = 1
N = 50

X = zeros(2, N^2)
X[1,:] = rand(N^2) * W
X[2,:] = rand(N^2) * H;

DM = DiffMap(X, m=20, Ïƒ=0.1);
```

```julia
p = scatter(X[1,:], X[2,:], zcolor=X[2,:], markersize=5, markerstrokewidth=0.3, markerstrokecolor=:black, 
    xlabel="\$x_1\$", ylabel="\$x_2\$", legend=:none, aspect_ratio = 1)
savefig("imgs/presentation-rectangle.png")
p
```

```julia
p = scatter(DM.Y[1,:], DM.Y[2,:], zcolor=X[2,:], markersize=5, markerstrokewidth=0, markerstrokecolor=:black, 
    xlabel="\$\\phi_1\$", ylabel="\$\\phi_2\$", legend=:none, aspect_ratio = 1)
savefig("imgs/presentation-rectangle1.png")
p
```

```julia
p = scatter(DM.Y[1,:], DM.Y[2,:], zcolor=X[2,:], markersize=5, markerstrokewidth=0, markerstrokecolor=:black, 
    xlabel="\$\\phi_1\$", ylabel="\$\\phi_2\$", legend=:none)
savefig("imgs/presentation-rectangle1b.png")
p
```
