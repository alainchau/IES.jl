```julia
using Plots
gr()
using IES
using NearestNeighbors
using 
```

```julia
W = 6
H = 1
N = 30

# Dense sample of grid
X = zeros(2, N^2)
X[1,:] = rand(N^2) * W
X[2,:] = rand(N^2) * H;
```

```julia
DM = DiffMap(X, m=20, Ïƒ=0.1);
```

```julia
idx = 50 # Index of point under consideration

Y = DM.Y

kdt = KDTree(X)
nn = knn(kdt, X[:,idx], 20)[1]

kdt2 = KDTree(Y)
nn2 = knn(kdt2, Y[:,idx], 20)[1]

p = scatter(Y[1,:], Y[6,:], markersize=2, legend=:none)
scatter!(Y[1, nn], Y[6, nn], markersize=8, markercolor=:red, markershape=:circle, legend=:none)
scatter!(Y[1, nn2], Y[6, nn2], markersize=4, markershape=:diamond, markercolor=:lightblue, legend=:none)
scatter!(Y[1, idx:idx], Y[6,idx:idx], markersize=8, markercolor=:yellow)
xlims!(-0.0005, -0.0003)
ylims!(-0.005, 0.005)
savefig("imgs/neighbors.png")
p
```
