```julia
using Revise
using Plots
# plotly()
gr()
using IES
using StatsBase
using LinearAlgebra
```

```julia
function dataset(p=10, N=500)

    X = zeros(p, N)
    X[1,:] = 6*rand(N)
    X[2,:] =   rand(N)
    cc = deepcopy(X[2,:])

    # Rotate the rectangular strip
    function randRotation(p)
        R = diagm(ones(p))
        i,j = samplepair(1:p)
        θ = 2π*rand()
        R[i,i] = cos(θ)
        R[i,j] = -sin(θ)
        R[j,i] = sin(θ)
        R[j,j] = cos(θ)
        R
    end

    for _ in 1:500
        X = randRotation(p) * X
    end
    X
end
```

```julia
times = Dict(:br => [], :ag => [], :sg => [])
n = 500

ζ = 0.1
p = 3
slst = [2,3,4,5,6,7,8,9,10]
for s in slst
    X = dataset(p, n)
    DM = DiffMap(X, m=20, σ=0.2);
    mfd = rmetric(DM, d=2);
    
#     tic = time()
#     S = ies(mfd, :br, s=s, ζ=ζ)
#     println(S)
#     push!(times[:br], time() - tic)
    
    tic = time()
    S = ies(mfd, :ag, s=s, ζ=ζ)
    println(S)
    push!(times[:ag], time() - tic)

    tic = time()
    S = ies(mfd, :sg, s=s, ζ=ζ)
    println(S)
    push!(times[:sg], time() - tic)
end

plot(slst, times[:ag], label="AG", legend=:topleft, xlabel="s (dimension of embedding)", ylabel="time (secs)")
plot!(slst, times[:sg], label="SG")
# plot!(slst, times[:br])
```

```julia
X = dataset(20, 2500)
@time DM = DiffMap(X, m=20, σ=0.2);
```

```julia
@time mfd = rmetric(DM, d=2);
```

```julia
function plotdm(DM, S, cc)
    scatter(DM.Y[S[1],:], DM.Y[S[2],:], zcolor=cc, markersize=4, markerstrokewidth=0, markerstrokecolor=:black, 
        xlabel="\$\\phi_$(S[1])\$", ylabel="\$\\phi_$(S[2])\$", legend=:none)

end


plotdm(DM, [1,2], cc)
```
